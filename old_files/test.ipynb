```python
to_update = data.copy()
to_update['SKU FULL'] = to_update['SKU FULL'].apply(lambda x: '-'.join(x.split('-')[:-1]))
to_update
```

```python
to_add = others[others['Sku Styleisnow'].isin(to_update['SKU FULL'].values.tolist())]
to_add.head(2)
```

```python
for _, row_updated in to_update.iterrows():
    print(row_updated)
    
    print(row_updated['SKU FULL'])
    idx = to_add[to_add['Sku Styleisnow'] == row_updated['SKU FULL']]
    print(idx)
    
    for _, row_others in to_add.iterrows():
        qty, sizes = row_others['Qty Detail'], [x.upper() for x in row_others['Size']]
        print(qty, sizes)
        break
    break        
```

```python
to_add['Size'] = to_add['Size'].str.split(',')
to_add['Qty Detail'] = to_add['Qty Detail'].str.split(',').apply(lambda x: list(map(int, x)))

# Update the quantities based on to_update
for idx, row in to_add.iterrows():
    # Convert the 'Size' and 'Qty Detail' to a list if not already done
    sizes = row['Size']
    qty_details = row['Qty Detail']
    
    # Iterate over the rows in to_update to find and update the corresponding size
    for _, update_row in to_update.iterrows():
        if str(update_row['size']) in sizes:
            size_index = sizes.index(str(update_row['size']))  # Find the index of the size
            qty_details[size_index] = update_row['qty']  # Update the quantity

    # Update the DataFrame with the new qty details
    to_add.at[idx, 'Qty Detail'] = ','.join(map(str, qty_details))
    to_add.at[idx, 'Size'] = ','.join(map(str, sizes))
```
